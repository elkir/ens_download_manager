```python
Path().cwd()
filename = Path("../lists/mars_v02a_2017-07-13.list")
filename,filename.exists()
```

```python
df = pd.read_table(filename, header=14,skipfooter=6,delim_whitespace=True,
                   dtype={'file': 'int',
                        'length': 'int',
                        'missing': 'str',
                        'number': 'int',
                        'offset': 'int',
                        'param': 'str',
                        'step': 'int'}
                   )
df
```
